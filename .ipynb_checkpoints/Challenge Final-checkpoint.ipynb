```python
def daily_entry_turnstile(df):
    listz = []  
    
    for category,dataframe in list(df.groupby(['C/A','UNIT','SCP','STATION', 'DAY'])):
        dataframe['Entered'] = dataframe['ENTRIES'] - dataframe['ENTRIES'].shift()
        dataframe['ENTRIES'] = dataframe['Entered'].sum()
        listz.append(dataframe)
        
    return pd.concat(listz, ignore_index=True)
```

```python
df = daily_entry_turnstile(df)
df.head()
```
