```python
df = pd.read_csv('turnstile_160917.csv')

df.rename(columns = {'C/A' : 'CA'}, inplace=True)
df.rename(columns = {'EXITS                                                               ':'EXITS'}, inplace = True)
df.head()
```

```python
df2 = pd.DataFrame()
df2['Keys'] = tuple(zip(df.CA, df.UNIT, df.SCP, df.STATION))
# df2['Date'] = pd.to_datetime(df['DATE'],infer_datetime_format=True)
df2['DateTime'] = pd.to_datetime(df['TIME'],infer_datetime_format=True)
df2['Day'] = df2.DateTime.map(lambda x: x.day)
df2['Time'] = df2.DateTime.map(lambda x: x.hour)
```

```python
df2.head()
```

```python
df2['EntryCount'] = df.ENTRIES
df2['ExitCount'] =df.EXITS
```

```python
 df2['entries']= df2.groupby(['Keys'])['EntryCount'].diff()
df2['exits'] = df2.groupby(['Keys'])['ExitCount'].diff()
df2.head(8)
```

```python
df2['Impressions'] = df2.entries + df2.exits
df2.head(15)
```
