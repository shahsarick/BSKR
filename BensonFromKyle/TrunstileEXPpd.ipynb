```python
df = pd.read_csv('turnstile_160917.csv')
df.head()
```

```python
df.rename(columns = {'C/A' : 'CA'}, inplace=True)
```

```python
# displays all turnstiles for 59th st station at 12AM on 9/10/2016
df.loc[(df.STATION == '59 ST')&(df.TIME == '00:00:00')&(df.DATE == '09/10/2016')]
```

```python
# this confirms that any 'C/A' will only have one 'UNIT'  
u_units = df.groupby(['CA']).UNIT.nunique()
for x in u_units:
    if x!=1:
        print(x)
#  returns no results- x is always 1   
```

```python
# tells us any unit can have multiple CA's 
u_ca = df.groupby(['UNIT']).CA.nunique()
for y in u_ca:
    if y != 1:
        print(y)
```

```python
# for each station, only 1 unique CA, UNIT pair exists
df.groupby(['STATION','CA']).UNIT.nunique()
```
